# Fetch data

```python
import yfinance as yf
```

```python
# Replace 'AAPL' with the other stock symbols

stock_ticker = 'AAPL'


# Fetch data in 1y time frame (can be changed)

stock_data = yf.download(stock_ticker, start='2023-01-01', end='2024-01-01')
stock_data_long = yf.download(stock_ticker, start='2000-01-01', end='2024-01-01')


# Display the data
print(stock_data.head())  # Print the first few rows of the retrieved data
```

# Visualize Closing Stock price

```python
import matplotlib.pyplot as plt
```

```python
# Plotting the closing price
plt.figure(figsize=(10, 6))
stock_data['Close'].plot()
plt.title(f"{stock_ticker} Stock Price")
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.grid(True)
plt.show()
```

# Moving average

```python
# Calculate 50-day moving average

stock_data_long['MA_50'] = stock_data_long['Close'].rolling(window=50).mean()

# Calculate 50-day moving average

stock_data_long['MA_200'] = stock_data_long['Close'].rolling(window=200).mean()


# Define start and end dates
start_date = '2016-01-01'
end_date = '2022-12-31'

# Slice the DataFrame based on the specified date range
stock_data_subset = stock_data_long.loc[start_date:end_date]


# Plotting the closing price and moving average

plt.figure(figsize=(10, 6))
stock_data_subset[['Close', 'MA_50','MA_200']].plot()
plt.title(f"{stock_ticker} Stock Price with 50-day Moving Average & 200-day Moving Average")
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.grid(True)
plt.show()
```

# Implement a simple (maybe stupid trading strategy)

If the MA 50 and MA 200 cross and  MA 50 > MA 200 we buy stock  and if MA 50 and MA 200 cross and  MA 50 < MA 200 we sell stock

```python
# Identify crosses


# Define start and end dates
start_date = '2022-01-01'
end_date = '2024-01-01'

# Slice the DataFrame based on the specified date range
stock_data_subset = stock_data_long.loc[start_date:end_date]


cross_buy = (stock_data_subset['MA_50']>stock_data_subset['MA_200']) & (stock_data_subset['MA_50'].shift(1) < stock_data_subset['MA_200'].shift(1))
cross_sell = (stock_data_subset['MA_50']<stock_data_subset['MA_200']) & (stock_data_subset['MA_50'].shift(1) > stock_data_subset['MA_200'].shift(1))


# Plotting Buy and Sell Signals (Golden Cross and Death Cross)
plt.plot(stock_data_subset.index[cross_buy], stock_data_subset['Close'][cross_buy], '^', markersize=10, color='g', label='Buy Signal')
plt.plot(stock_data_subset.index[cross_sell], stock_data_subset['Close'][cross_sell], 'v', markersize=10, color='r', label='Sell Signal')


# Plotting MA 50 and MA 200
plt.plot(stock_data_subset['MA_50'], label='MA 50')
plt.plot(stock_data_subset['MA_200'], label='MA 200')

plt.title(f"{stock_ticker} Stock Price with Moving Averages and Crossovers")
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)

plt.show()
```

# Daily Returns

```python
# Calculate Daily returns

stock_data['Daily_Return'] = stock_data['Close'].pct_change()*100

# Plotting daily returns
plt.figure(figsize=(10, 6))
stock_data['Daily_Return'].plot()
plt.title(f"{stock_ticker} Daily Returns in %")
plt.xlabel('Date')
plt.ylabel('Daily Return')
plt.grid(True)
plt.show()

#Largest return and largest loss

print(f" The largest daily return was {round(stock_data['Daily_Return'].max(),2)}% on {stock_data['Daily_Return'].idxmax().strftime('%Y-%m-%d')} ")
print(f" The largest daily loss was {round(stock_data['Daily_Return'].min(),2)}% on {stock_data['Daily_Return'].idxmin().strftime('%Y-%m-%d')} ")
```
